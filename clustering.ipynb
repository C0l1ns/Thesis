```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("./datasets/train_energy_eff_Y1.csv").drop(['Unnamed: 0'],axis=1).to_numpy()
test_data = pd.read_csv("./datasets/test_energy_eff_Y1.csv").drop(['Unnamed: 0'],axis=1).to_numpy()
```

```python
X_train, y_train = train_data[:,:-1], train_data[:,-1:]
X_test, y_test = test_data[:,:-1], test_data[:,-1:]
```

```python
# sc = StandardScaler()
# scaled = sc.fit_transform(X)
pca = PCA(n_components=2)
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters,random_state=22, n_init="auto").fit(train_data)
```

```python
pca_data = pd.DataFrame(pca.fit_transform(train_data),columns=['PC1','PC2']) 
pca_data['cluster'] = pd.Categorical(kmeans.labels_)
sns.scatterplot(x="PC1",y="PC2",hue="cluster",data=pca_data)
```

```python
# sc = StandardScaler()
# scaled = sc.fit_transform(X_train)
pca = PCA(n_components=2)
kmeans = KMeans(n_clusters=n_clusters,random_state=22, n_init="auto").fit(X_train)
pca_data = pd.DataFrame(pca.fit_transform(X_train),columns=['PC1','PC2']) 
pca_data['cluster'] = pd.Categorical(kmeans.labels_)
sns.scatterplot(x="PC1",y="PC2",hue="cluster",data=pca_data)
```
